```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cleaned_dataset.csv")
```

```python
data.head(3)
```

```python
# Realtion between stake and gain, by bet_type
sns.scatterplot(data=data, x='stake_log', y='gain_log', hue='bet_type', alpha=0.6)
plt.title('Stake vs Gain by Bet Type')
plt.xlabel('Stake')
plt.ylabel('Gain')
plt.legend(title='Bet Type')
plt.show()

```

```python
# many results in 0 gain, high density along horizontal base.
# gain inreases with stake, trend from the left to right
# blue dots(multiple bet_type) higher than orange
```

```python
# Realtion between GGR and odds, by sport
sns.scatterplot(data=data, x='odds_log', y='GGR_log', hue='sport', alpha=0.6)
plt.title('odds vs GGR by sport')
plt.xlabel('Odds')
plt.ylabel('GGR')
plt.legend(title='sport')
plt.show()
```

```python
# color overlapping indicate similar odds ranges.
# Betting activity occurs at moderate odds with moderate revenue outcome.
```

```python
sns.lmplot(data=data, x='odds_log', y='GGR_log', hue='sport', scatter=False, height=6, aspect=1.5, ci=None)
plt.title('Trend Lines: Odds vs GGR by Sport')
plt.xlabel('Odds')
plt.ylabel('GGR')
plt.show()
```

```python
# odds increase, GGR decrease. Small change inGGR as odds change: Formula 1 or Football.
# large engative impact on GGR as odds increase: Table tennis or Cycling.
```

```python
avg_stake_per_sport = data.groupby('sport')['stake_log'].mean().reset_index()
avg_stake_per_sport = avg_stake_per_sport.sort_values(by='stake_log', ascending=False)

# Plot line chart
sns.barplot(data=avg_stake_per_sport, x='stake_log', y='sport', palette='viridis')
plt.title('Average Stake per Sport')
plt.xlabel('Average Stake')
plt.ylabel('Sport')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# there are small difference in average stake, higher are Snooker and Formula 1.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='odds_log', y='stake_log', alpha=0.5)
plt.title('Stake vs Odds')
plt.xlabel('Odds')
plt.ylabel('Stake')
plt.grid(True)
plt.tight_layout()
plt.show()
```
